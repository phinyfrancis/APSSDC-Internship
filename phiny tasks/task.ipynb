```python
import operator
with open('files/task.csv','r') as f:
    s=f.read()
    st=""
    list1=[]
    list2=[]
    rollname=[]
    ts=[]
    for i in s:
        if(i=="\n"):
            list1.append(st.split(","))
            list2.append(list1)
            st=""
        else:
            st=st+i
    for i in list2:
        for j in i:
            ts.append(j[4])
            
    for i in list2:
        for j in i:
            rollname.append(j[0])
    rollname.pop(rollname.index("Name"))
    rollnolist=[]
    for i in rollname:
        if(operator.contains(i,'-')==False):
            pass
            
        else:
            no=i.split("-")
            count=0
            for j in no[0]:
                if(j.isnumeric()):
                    count+=1
            if(count>0):
                rollnolist.append(no[0])
    
            
            
            
            
        
            
            
        
    
            
        
            
            
    
    

    
    
    
```

```python
with open('files/output.csv','w') as w:
    s1=list(set(rollnolist))
    s2=list(set(ts))
    w.write("RollNo"+","+"Time"+"\n")
    for i in range(0,len(s1)):
        w.write(s1[i]+","+s2[i]+"\n")
    
```
